```python
#To use scikit-learn library, we have to convert the Pandas data frame to a Numpy array:

X=df.iloc[:, :-1].values
X
```

```python
y = df.iloc[:,13]
y
```

```python
#Splitting the dataset into Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
# scaling the dataset
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#Shape of train and test data
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Using KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Train Model  
neigh = KNeighborsClassifier(n_neighbors = 4).fit(X_train,y_train)
```

```python
#Prediction
prediction = neigh.predict(X_test)
```

```python
prediction1=pd.DataFrame(prediction)
prediction1.head()
```

```python
#Accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, prediction)
```

```python
# Now Using SVM Algorithm 
from sklearn.svm import SVC
classifier=SVC(kernel='rbf',random_state=0)
classifier.fit(X_train,y_train)
```

```python
#Predict
y_pred=classifier.predict(X_test)
```

```python
prediction2=pd.DataFrame(y_pred)
prediction2.head()
```

```python
#Accuracy
metrics.accuracy_score(y_test, prediction2)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction1)
```

```python
confusion_matrix(y_test,prediction2)
```

```python
#Using DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
HeartTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
```

```python
HeartTree.fit(X_train,y_train)
```

```python
y_pred = HeartTree.predict(X_test)
```

```python
prediction3=pd.DataFrame(y_pred)
prediction3.head()
```
